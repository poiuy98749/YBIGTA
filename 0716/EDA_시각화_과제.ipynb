### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
missing_columns = df.isnull().sum() / len(df) * 100
# print("Missing columns.shape:", missing_columns.shape)
missing_columns = missing_columns[missing_columns > 0]
# print("Missing columns.shape:", missing_columns.shape)

missing_columns_df = pd.DataFrame(missing_columns, columns=['missing %'])
missing_columns_df = missing_columns_df.sort_values(by='missing %',
                                                    ascending=False)
missing_columns_df
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize=(15,5)) # Set figure size
plt.xticks(rotation=45)
# sns.set_palette("crest")

sns.barplot(data=missing_columns_df, x=missing_columns_df.index, y='missing %')
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다. 
  #### <font color='orange'> 'beds' 컬럼이 결측치인 데이터의 상위 5개를 출력해주세요.

```python
# Q4
df[df['beds'].isnull()][:5]
```

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

```python
# Q6
frequency = df['room_type'].value_counts()
percentage = frequency / sum(frequency)

ax = sns.barplot(list(percentage.index), list(percentage))
ax.set(ylabel = "room_type")
```
