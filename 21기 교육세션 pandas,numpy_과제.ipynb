#### 'strength' 컬럼추가 : alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (상위 30개만 출력)

```python
def getStrength(row):
    if row['alcohol'] > 10:
        return 'strong'
    else:
        return 'weak'

data['strength'] = data.apply(lambda row: getStrength(row), axis=1)
data[:30]
```

#### merge 실습을 위한 데이터 쪼개기


```python
df_x = data[['quality', 'fixed acidity', 'citric acid']].head(30)
```

```python
df_y = data[['quality', 'volatile acidity', 'chlorides', 'alcohol']].head(30)
```

```python
df_y = df_y[df_y.quality >= 7]
```

```python
len(df_y)
```

#### df_x와 df_y를 inner join시켜라 (quality를 기준으로)

```python
df_innerJoin = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='inner')
df_innerJoin
```

#### df_x에서 df_y를 left join시켜라 (quality를 기준으로)

```python
df_leftJoin = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='left')
df_leftJoin
```

# <과제2>
- 구글링도 과제 수행에 포함되어 있습니다. 세션에 소개되지 않은 기능이 필요하다면, 구글링을 통해 구현해주시길 바랍니다.
- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
data = pd.read_csv('deliver.csv') #wine.csv도 '내드라이브'에 사본 저장 후, 경로 수정
data.head()
```

```python
# 월, 계절 추가
def getMonth(row):
    SEASON_LIST = ['EMPTY', '겨울', '겨울', '봄', '봄', '봄', '여름', '여름', '여름', '가을', '가을', '가을', '겨울']
    date = row['날짜']
    # print(date)
    year, month, day = list(map(int, date.split('-')))
    
    
    return str(month), SEASON_LIST[month]


data['월'] = data.apply(lambda row: getMonth(row)[0], axis=1)
data['계절'] = data.apply(lambda row: getMonth(row)[1], axis=1)

data.head()
```

```python
# Null값 확인 후 대체
data['미세먼지(㎍/㎥)'].isnull().sum()

data['미세먼지(㎍/㎥)'] = data['미세먼지(㎍/㎥)'].fillna((data['미세먼지(㎍/㎥)'].mean()))

data['미세먼지(㎍/㎥)'].isnull().sum()
```
